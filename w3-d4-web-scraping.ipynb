```python
tables = soup.find_all("table")
len(tables)
```

```python
print(type(tables[0]))
tables[0].get("class")
```

```python
[table.get("class") for table in tables]
```

```python
medallero = tables[10]
medallero
```

```python
lineas = medallero.find('tbody').find_all("tr")[1:]
```

```python
lineas
```

```python

def structureLine(tag):
    td = tag.find_all("td")
    return {
        "orden": td[0].text,
        "pais":  td[1].find("a").text,
        "oro": int(td[2].text),
        "plata": int(td[3].text),
        "bronce": int(td[4].text),
    }

structureLine(lineas[0])
```

```python
medalleroList = list(map(structureLine, lineas))
medalleroList
```

```python
import pandas as pd
```

```python
medallero_df = pd.DataFrame(medalleroList)
medallero_df
```

```python
medallero_df.oro.sum()
```

# Web de deliveroo

```python
html = requests.get("https://deliveroo.es/es/restaurants/madrid/legazpi").text
soup = BeautifulSoup(html)
```

```python
#restaurantes = soup.select("div[class^=HomeFeedUICard]")
restaurantes = soup.select(".HomeFeedUICard-9e4c25acad3130ed")

```

```python
def processRestaurant(rest):
    try:
        name = rest.select_one("li p").text
        return {
            "name":name,
            "valoracion":rest.select(".HomeFeedUICard-45fe00a3d559fba5 li span.ccl-19882374e640f487")[0].text
        }
    except:
        return
```

```python
processRestaurant(restaurantes[0]) 
```

```python
restaurantsStructured = [processRestaurant(rest) for rest in restaurantes]
restaurantsStructured
```

```python
restaurantsStructured = list(filter(lambda r: type(r)==dict, restaurantsStructured))
restaurantsStructured
```

```python
def valora(rest):
    try:
        val = float(rest["valoracion"].split(" ")[0])
    except:
        val = 0.0
    return {"name":rest["name"], "valoracion": val}
restaurantsStructured = list(map(valora,restaurantsStructured))
```

```python
import pandas as pd
pd.DataFrame(restaurantsStructured)
```
