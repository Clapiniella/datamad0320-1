## FFT

```python
Fm = 1000
X = np.linspace(0,1,Fm)

noise = 0.2* np.sin(2*np.pi*50*X)
real = np.sin(2*np.pi*1*X)
Y = noise + real
plt.plot(X,Y)
plt.plot(X,real)
plt.legend(["Measured","Real"])
```

```python
from scipy.fftpack import fft
fft_y = np.abs(fft(Y))
plt.plot(fft_y[0:500])
```

## Applying a low pass butterworth filter

```python
from scipy import signal

# Definimos la frecuencia de corte del filtro
Fc=10
b, a = signal.butter(3, Fc/Fm)

# aplicamos el filtro
Y_clean = signal.filtfilt(b, a, Y)
```

```python
plt.plot(X,Y)
plt.plot(X,Y_clean)
plt.legend(["Measured","Cleaned"])
```

```python
fft_y = np.abs(fft(Y_clean))
plt.plot(fft_y[0:500])
```

## Cleaning temperature data

```python
# Definimos la frecuencia de corte del filtro
Fc=20
b, a = signal.butter(3, Fc/Fm)
temp_clean = pd.Series(signal.filtfilt(b, a, noisy_temp.values), index=noisy_temp.index)
```

```python
plt.plot(noisy_temp.loc[analysis_range].values)
plt.plot(temp_clean.loc[analysis_range].values)
plt.legend(["noise_temp","clean_temp"])

plt.figure()
plt.plot(temp_clean.loc[analysis_range].values)
```

# Predicting time series data

```python
from statsmodels.tsa.ar_model import AR

# T = (a*(T-1) + b*(T-2))

#22 = a*21 + b*22
#23 = a*21.5 + b*23.4
```

```python
train = temp_clean[:700] ## con lo que entrenamos el modelo
test = temp_clean[700:] ## con lo que evaluamos el modelo
print(train.index)
```

```python
model = AR(train)
fit_model = model.fit(maxlag=4)
fit_model.params
```

```python
y_pred = fit_model.predict(start=len(train),end=len(train)+len(test)-1, dynamic=False)
```

```python
from sklearn.metrics import mean_squared_error

results = pd.DataFrame({
    "GT":test,
    "pred":y_pred,
    "diff":y_pred-test
})
print("MSE = ",round(mean_squared_error(test,y_pred),5))
print("RMSE = ",np.sqrt(mean_squared_error(test,y_pred)))
display(results)
```

# 
ARMA Model

```python
from statsmodels.tsa.arima_model import ARMA

analysis_range = pd.date_range('2015-02-15', periods=24*30,freq="H")
co2 = occupancy.CO2.loc[analysis_range]
train = co2[:700]
test = co2[700:]
```

```python
model = ARMA(train, order=(2,2))
arma_results = model.fit()
```

```python
y_pred = arma_results.predict(start=len(train),end=len(train)+len(test)-1, dynamic=False)
```

```python
results = pd.DataFrame({
    "GT":test,
    "pred":y_pred,
    "diff":y_pred-test
})
print("MSE = ",round(mean_squared_error(test,y_pred),5))
print("RMSE = ",np.sqrt(mean_squared_error(test,y_pred)))
display(results)
```
