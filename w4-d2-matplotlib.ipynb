```python
def square(L=1):
    square = np.asarray([
        [0,0],
        [0,L],
        [L,L],
        [L,0],
        [0,0]
    ])
    return square.T

plt.figure(figsize=(5,5))
plt.plot(*square(5),'o-')
plt.plot(*square(10),'o-')
```

```python
def square(L=1, c=(0,0)):
    square = np.asarray([
        [0,0],
        [0,L],
        [L,L],
        [L,0],
        [0,0]
    ])
    return (square - c - L/2).T

plt.figure(figsize=(5,5))
#square(5)

plt.plot(0,0,'o')
plt.plot(*square(5),'o-')
plt.plot(*square(10),'o-')
```

```python
# https://es.wikipedia.org/wiki/Matriz_de_rotaci%C3%B3n
def square(L=1, c=(0,0), angle=0):
    angle = np.pi*angle/180
    R = np.asarray([
        [np.cos(angle), -np.sin(angle)],
        [np.sin(angle), np.cos(angle)]
    ])
    
    square = np.asarray([
        [0,0],
        [0,L],
        [L,L],
        [L,0],
        [0,0]
    ])
    return np.dot((square - c - L/2),R).T

plt.figure(figsize=(5,5))
#square(5)

plt.plot(0,0,'o')
plt.plot(*square(5, angle=45),'o-')
```

```python
plt.figure(figsize=(5,5))

with plt.xkcd():
    for angle in np.linspace(0,180,10):
        plt.plot(*square(5, angle=angle),'o-')
```

```python
plt.figure(figsize=(5,5))
for idx,angle in enumerate(np.linspace(0,180,20)):
    plt.plot(*square(idx+10, angle=angle),'o')
```

# Ploting with pandas

```python
import pandas as pd
df = pd.DataFrame(np.random.random((12,2))*1000, columns=["TiendaA","TiendaB"])
df.index = [f"Mes{i}" for i in range(12)]

df.plot.bar()
plt.title("Ventas por ubicación")
plt.xlabel("Meses")
plt.ylabel("Ventas (€)")
```

```python
import datetime
df = pd.DataFrame(np.random.random((12,2))*1000, columns=["TiendaA","TiendaB"])

df.plot.bar()
plt.title("Ventas por ubicación")
plt.xlabel("Meses")
plt.ylabel("Ventas (€)")


labelPos = np.arange(12)
tickLabel = [datetime.datetime(year=2020, month=i+1,day=1).strftime("%b %Y") for i in range(12)]
plt.xticks(labelPos,tickLabel, rotation=90)


plt.annotate("Nuevo Producto", xy=(0,df.iloc[0].max()), xytext=(3,df.max().max()), arrowprops=dict(facecolor='red', shrink=0.05))

plt.show()
```
