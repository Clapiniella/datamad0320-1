```python
import seaborn as sns
import pandas as pd

xz = np.concatenate([np.random.randn(1000)+10,np.random.randn(1000)])
yz = np.concatenate([np.random.randn(1000),np.random.randn(1000)])

df = pd.DataFrame({"x":xz,"y":yz})
sns.jointplot(x="x", y="y", data=df, kind="kde");
```

```python
xz = np.concatenate([np.random.randn(1000)+2,np.random.randn(1000)-2])
yz = np.concatenate([np.random.randn(1000)+2,np.random.randn(1000)-2])

df = pd.DataFrame({"x":xz,"y":yz})
sns.jointplot(x="x", y="y", data=df, kind="kde");
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
data_2d = pca.fit_transform(X)
```

```python
print(data_2d.shape)

df = pd.DataFrame({"x":data_2d[:,0],"y":data_2d[:,1]})
sns.jointplot(x="x", y="y", data=df, kind="kde");
```

```python
idx5 = np.where(y=='5')
idx8 = np.where(y=='8')
print(idx5[0].shape,idx8[0].shape)
```

```python
five = X[idx5]
```

```python
for i,x in enumerate(five):
    plt.figure()
    plt.imshow(x.reshape((28,28)), cmap="gray")
    if i>5:
        break
```

```python
five = data_2d[idx5]
eight = data_2d[idx8]

plt.scatter(five[:,0],five[:,1],alpha=0.2)
plt.scatter(eight[:,0],eight[:,1],alpha=0.2)
```

```python
def getFeaturesFrom(n):
    return data_2d[np.where(y==f"{n}")].T

numbers=[4,0]

for number in numbers:
    plt.scatter(*getFeaturesFrom(number),alpha=0.2)

plt.legend(numbers)
```
