```python
import requests
from bs4 import BeautifulSoup

url = "https://www.emsc-csem.org/Earthquake/"
data = requests.get(url).content.decode('iso_8859_1')
print(data)
#.encode('iso_8859_1')
soup = BeautifulSoup(data)
soup
```

```python
table = soup.select_one("#tbody")
lines = table.find_all("tr")
lines[0]
```

```python
import datetime

def processLine(line):
    try:
        latlng = list(map(lambda tag: float(tag.get_text().strip()), line.select(".tabev1")))
        fecha = line.select_one(".tabev6").find("a").text.replace('\xa0',' ').split('.')[0].split(' ')
        fecha = [int(e) for e in fecha[0].split('-') + fecha[-1].split(":")]
        date = datetime.datetime(*fecha)
        return {
            "lat":latlng,
            "lng":latlng,
            "date": date.isoformat(),
            "depth": int(line.select_one(".tabev3").text),
            "region":line.select_one(".tb_region").text.strip()
        }
    except:
        return

processLine(lines[0])

```

```python
import pandas as pd
pd.DataFrame([processLine(line) for line in lines])
```
